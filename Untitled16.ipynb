```python
#Train_Test_split
#into X_train,X_test,Y_train,Y_train
```

```python
X=heartds.drop(columns='target',axis=1)
```

```python
Y=heartds['target']
```

```python
Y
```

```python
#Splitting into train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
#Using random state data is split randomly
```

```python
#MODEL TRAINING
#USING LOGISTIC REGRESSION
```

```python
#we are going to use logistic regression
model = LogisticRegression()
```

```python
#training our model with train data
```

```python
model.fit(X_train, Y_train)
```

```python
#evaluationg our model 
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Train Data:',training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data:',testing_data_accuracy)
```
